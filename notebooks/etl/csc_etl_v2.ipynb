```python
def rebuild_file(file_grp):
    rows = []
    header = []
    n_cols = []
        
    for i, file in tqdm(enumerate(file_grp), total=len(file_grp)):
        
        with open(file, 'r') as f:
            
            
            lines = []
            for i, l in enumerate(f):
                for bid in bad_ids:
                    if bid in l:
                        print(f'Found {bid} in {file} line {i+1}')
                        
                row = [s.strip() for s in l.split('|')]
                n_cols.append(len(row))
                if len(n_cols) > 1 and n_cols[-1] != n_cols[-2]:
                    print(f'Messed up columns in {file} line {i+1}')
                    print(l)
                    print(row)
                    break
#                     raise Exception("Parsing error")
                lines.append(row)
            
            if i == 0 or lines[0] == header:
                header = lines[0]
                rows = lines[1:]
            else:
                rows += lines
        if len(n_cols) > 1 and n_cols[-1] != n_cols[-2]:
            break
                
    return pd.DataFrame(rows, columns=header)
```

```python
def rebuild_file(file_grp):
    rows = []
    header = []
    n_cols = []
        
    for i, file in tqdm(enumerate(file_grp), total=len(file_grp)):
        
        with open(file, 'r') as f:
            
            
            lines = []
            for i, l in enumerate(f):
                for bid in bad_ids:
                    if bid in l:
                        print(f'Found {bid} in {file} line {i+1}')
                        
                row = [s.strip() for s in l.split('|')]
                n_cols.append(len(row))
                if len(n_cols) > 1 and n_cols[-1] != n_cols[-2]:
                    print(f'Messed up columns in {file} line {i+1}')
                    print(l)
                    print(row)
                    break
#                     raise Exception("Parsing error")
                lines.append(row)
            
            if i == 0 or lines[0] == header:
                header = lines[0]
                rows = lines[1:]
            else:
                rows += lines
        if len(n_cols) > 1 and n_cols[-1] != n_cols[-2]:
            break
                
    return pd.DataFrame(rows, columns=header)
```
